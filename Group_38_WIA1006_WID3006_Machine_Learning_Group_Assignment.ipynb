````python
import numpy as np
import seaborn as sbn
import matplotlib.pyplot as plt

def apply_log_transform(data, skewness_threshold=1):
    log_transformed_data = data.copy()

    # Apply log transformation to columns with skewness above the threshold
    for col in log_transformed_data.columns:
        skewness = log_transformed_data[col].skew()
        if abs(skewness) > skewness_threshold:
            # Find the minimum value in the column
            min_val = log_transformed_data[col].min()

            # If the minimum value is less than or equal to 0, shift all values in the column
            if min_val <= 0:
                log_transformed_data[col] += abs(min_val) + 1

            # Apply log transformation
            log_transformed_data[col] = np.log10(log_transformed_data[col])

    return log_transformed_data

# Example usage
df_numerical_log = apply_log_transform(df_numerical, skewness_threshold=1)

# Plot boxplot for all log-transformed data
plt.figure(figsize=(16, 8))
sbn.boxplot(data=df_numerical_log, orient="h")
plt.title('Log-transformed Data')
plt.show()
````

Even after winsorization, all numerical variables are still hgihly skewed (abs skewness > 1). Thus we decided to apply log transformation to the data. As we can see from the boxplot, most features no longer have falsely marked outliers. However, some features still have outliers that could be explained by the sheer difference in financial metrics like 'Net Income' and 'Total Receivables' between small and giant companies.

````python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest

# Assuming df_numerical_log is your log-transformed numerical DataFrame
# Initialize Isolation Forest with auto contamination
iso_forest = IsolationForest(contamination=0.02, random_state=42)
outlier_df = df_numerical_log.copy()

# Fit the model
iso_forest.fit(outlier_df)

# Predict outliers
outliers = iso_forest.predict(outlier_df)
# -1 for outliers, 1 for inliers
outliers = pd.Series(outliers, index=outlier_df.index)
outliers = outliers == -1

# Print number of outliers detected
print("Number of outliers detected:", outliers.sum())

# Impute outliers with the median of each column
df_numerical_imputed = outlier_df.copy()
for column in df_numerical_imputed.columns:
    median_value = df_numerical_imputed[column].median()
    df_numerical_imputed.loc[outliers, column] = median_value

print("Outliers imputed with median.")
````

- Financial data can have significant variations and extreme values due to differences in company sizes and financial capabilities. Traditional outlier detection methods like IQR and MAD can incorrectly categorize these extreme but genuine values like large conglomerates as outliers.

- Isolation Forest is an effective method in this scenario because it isolates outliers instead of profiling normal data points. It works by randomly selecting a feature and then randomly selecting a split value between the maximum and minimum values of the selected feature. The logic is that isolating anomaly observations is easier because only a few conditions are needed to separate cases that are drastically different from the rest.

- When it comes to imputing outliers, the median is a robust measure that is not influenced by outliers or extreme values, unlike the mean.

- This way, we are able to preserve important information in genuine extreme values for our model.

````python
# Plot boxplot for imputed data
plt.figure(figsize=(16, 8))
sbn.boxplot(data=df_numerical_imputed, orient="h")
plt.title('Log-transformed Data (Outlier imputed)')
plt.show()
````

It is important to note that again, outliers marked by boxplot are inaccurate for financial data as explained before. Observing closely, we can see some extreme values are classified as outliers by the isolation forest model. This preserves important information in genuine extreme values while handling outliers at the same time.

#### Handling Data Imbalance

There are 3 popular ways to handle data imbalance using resampling:

* Random Undersampling: Removing random records from the majority class.
* Random Oversampling: Adding random duplicates from the minority class.
* Synthetic Data Undersampling / Oversampling: Group values in majority class to reduce data points or generate values within minority classs to increase data points.

In this case, we will be using a combination of Synthetic Data Oversampling and Synthetic Data Undersampling technique called [Synthetic Minority Oversampling Technique (SMOTE)](https://www.analyticsvidhya.com/blog/2020/10/overcoming-class-imbalance-using-smote-techniques/#h-smote-synthetic-minority-oversampling-technique) and [Edited Nearest Neighbour (ENN)](https://bobrupakroy.medium.com/edited-nearest-neighbors-enn-c446a15e4bbe), whose combination is known as the [SMOTE-ENN algorithm](https://towardsdatascience.com/imbalanced-classification-in-python-smote-enn-method-db5db06b8d50) which can be accessed from the `imblearn` library.

In simpler words for SMOTE-ENN techniques, SMOTE increases the number of samples in the minority class by generating synthetic data points, while clean the dataset by removing potentially noisy or ambiguous samples from the majority class. This helps to create a balanced and cleaner dataset by addressing both overfitting in the minority class and noise in the majority class.

````python
from imblearn.over_sampling import SMOTE
from collections import Counter

# Separate features and target variable
X = df_encoded.drop('Status', axis=1)
y = df_encoded['Status']

from imblearn.combine import SMOTEENN

# Create an instance of SMOTEENN with a specified random state
# Apply SMOTEENN resampling to both features (X) and the target variable (y)
X_resampled, y_resampled = SMOTEENN(random_state=161223).fit_resample(X, y)


# Combine resampled data into a new dataframe
df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
df_resampled['Status'] = y_resampled

# Display new class counts
print(Counter(y_resampled))
````

In our original dataset, we observed a class imbalance with 93.6% of companies being ‘alive’ and only 6.6% being ‘failed’. ####**Method 1:** Filter Methods

**Chi-Squared Feature Selection:**

- Focuses on the relationship between categorical target variables and categorical features.
- Top features include: Financial Year, EBITDA, Total LTD, Retained Earnings, EBIT, etc.
- Strength: Identifies features with the strongest statistical association with the target.



**ANOVA Feature Selection:**

- Measures the difference in means of the target variable across different feature values.
- Top features include: Financial Year, Total LTD, EBIT, MajorGroup, Division, etc.
- Strength: Useful for continuous and normally distributed data.


**Linear Discriminant Analysis (LDA):**

- Projects features in a way that maximizes class separability.
- Top features include: Inventory, Total Revenue, EBITDA, Total assets, Dep & Amort, etc.
- Strength: Good for dimensionality reduction while retaining class discriminative information.

````python
# 1. Filter methods - SelectKBest with chi2
best_features_filter = SelectKBest(score_func=chi2, k=10)
fit = best_features_filter.fit(X, y)
features_scores = pd.DataFrame(fit.scores_)
features_columns = pd.DataFrame(X.columns)
feature_scores = pd.concat([features_columns, features_scores], axis=1)
feature_scores.columns = ['Feature', 'Score']
print("Chi-Squared Feature Selection:")
print(feature_scores.nlargest(10, 'Score'))

# ANOVA (f_classif)
best_features_filter_anova = SelectKBest(score_func=f_classif, k=10)
fit_anova = best_features_filter_anova.fit(X, y)
features_scores_anova = pd.DataFrame(fit_anova.scores_)
features_columns_anova = pd.DataFrame(X.columns)
feature_scores_anova = pd.concat([features_columns_anova, features_scores_anova], axis=1)
feature_scores_anova.columns = ['Feature', 'Score']
print("\nANOVA Feature Selection:")
print(feature_scores_anova.nlargest(10, 'Score'))

# Linear Discriminant Analysis (LDA)
n_components = min(X.shape[1], n_classes - 1)
lda = LinearDiscriminantAnalysis(n_components=n_components)
lda.fit(X, y)
selected_features_lda = pd.DataFrame({'Feature': X.columns, 'Score': lda.coef_[0]})
selected_features_lda = selected_features_lda.sort_values(by='Score', ascending=False).head(10)['Feature']
print("\nLinear Discriminant Analysis (LDA) Feature Selection:")
print(selected_features_lda)
````

####**Method 2:** Wrapper Methods

**Recursive Feature Elimination (RFE):**

- Recursively removes least important features based on model performance.
- Top features include: Current Assets, EBITDA, Inventory, Total LTD, EBIT, etc.
- Strength: Selects features that contribute most to the model's predictive power.


**Forward Selection:**

- Sequentially adds features that improve model performance.
- Selected features: Current Assets, Cost of Goods Sold, Dep & Amort, EBITDA, Inventory, etc.
- Strength: Simple and interpretable, good for smaller datasets.


**Backward Elimination:**

- Sequentially removes features that do not improve model performance.
Selected features: Current Assets, Dep & Amort, EBITDA, Inventory, Net Income, etc.
Strength: Can handle larger datasets and complex interactions.

````python
# 2. Wrapper methods - Recursive Feature Elimination (RFE) with Logistic Regression
model = LogisticRegression(max_iter=500, solver='saga')
rfe = RFE(estimator=model, n_features_to_select=10)
fit = rfe.fit(X, y)
selected_features_wrapper = pd.DataFrame({'Feature': X.columns, 'Rank': rfe.ranking_})
selected_features_wrapper = selected_features_wrapper[selected_features_wrapper['Rank'] == 1]['Feature']
print("\nRecursive Feature Elimination (RFE) Feature Selection:")
print(selected_features_wrapper)

X_df = pd.DataFrame(X, columns=X.columns)

# Forward Selection with Logistic Regression
sfs = SFS(LogisticRegression(max_iter=500, solver='saga'),
          k_features=10,
          forward=True,
          floating=False,
          scoring='accuracy',
          cv=5,
          n_jobs=-1)  # Use all available cores

# Fit the selector to the data
sfs.fit(X_df, y)

# Extract the selected feature names
selected_features_forward = list(X_df.columns[list(sfs.k_feature_idx_)])

print("\nForward Selection Feature Selection:")
print(selected_features_forward)


# Backward Elimination with Logistic Regression
sbs = SFS(LogisticRegression(max_iter=500, solver='saga'),
          k_features=10,
          forward=False,
          floating=False,
          scoring='accuracy',
          cv=5,
          n_jobs=-1)  # Use all available cores

# Fit the selector to the data
sbs.fit(X_df, y)

# Extract the selected feature names
selected_features_backward = list(X_df.columns[list(sbs.k_feature_idx_)])

print("\nBackward Elimination Feature Selection:")
print(selected_features_backward)
````

### Optimal Number of Features


````python
from sklearn.model_selection import cross_val_score
# Sample a subset of the data for feature selection
sample_fraction = 0.1  # Use 10% of the data
X_sampled = X.sample(frac=sample_fraction, random_state=42)
y_sampled = y.loc[X_sampled.index]

# Function to evaluate number of features
def evaluate_features(num_features):
    model = LogisticRegression(max_iter=500, solver='saga')
    rfe = RFE(estimator=model, n_features_to_select=num_features)
    X_rfe = rfe.fit_transform(X_sampled, y_sampled)  # Using pre-selected features on the sampled data
    scores = cross_val_score(model, X_rfe, y_sampled, cv=3, scoring='f1')  # Reduced to 3 folds
    return scores.mean()

# Evaluate different numbers of features
feature_counts = range(5, 23)
results = {n: evaluate_features(n) for n in feature_counts}

# Print the results
for n, score in results.items():
    print(f"Number of features: {n}, Cross-validated F1: {score:.4f}")

# Plot the results
plt.plot(list(results.keys()), list(results.values()), marker='o')
plt.xlabel('Number of Features')
plt.ylabel('Cross-validated F1')
plt.title('Optimal Number of Features')
plt.show()
````

Before we dive into modelling and hyperparameters tuning, please read the following to understand the necessary details:

1. Hyperparameter tuning - Grid Search
- Grid Search is a **hyperparameter tuning** technique used to find the best combination of hyperparameters for a given model.
- Define paramter grid by specifying the hyperparameters you want to tune and their respective values. This grid represents all the hyperparameters combination that will be tested.
- Use 'GridSearchCV' to pass the model, parameter grid, scoring metric and other parameters, then fit the grid object to training data and search through all the combinations of hyperparameter that give the best results based on specific scoring metric.

2. Performance evaluation metrics:
- **Accuracy**: measures the proportion of correct prediction out of the total number of predictions made by model
- **F1-score**: used to evaluate the performance of classification model particularly when the classes are imbalanced. It captures both precision and recall, best value at 1 and worst at 0.
- **Recall**: sensitivity or true positive rate, measures the ability of the classifier to correctly identify all the positive instances, best value at 1 and worst at 0.
- **ROC-AUC**: Evaluate the ability of binary classification model to differentiate between positive and negative classes. It is particularly useful in scenarios where the class is imbalanced.

3. Visualisation plot
- In this section, we mainly use heatmap to visualise three components: **model performance plot** across different hyperparameter settings, **confusion matrix** and **ROC-AUC graph**.
- By using Grid Search, we are able to quickly loop through all the combinations of hyperparameters and evaluate the model performance, the visual summary will help to identify which combinations of hyperparameters yield the best scores
- Confusion matrix tell us the value of True Positive(TP), False Positive(FP), True Negative(TN) and False Negative(FN). Basically it's a two dimensional matrix that tell us number of positive classes correctly predicted, number of positive classes wrongly predicted and vice versa.
- ROC-AUC graph provides a visual representation of the performance of a binary classification model across different threshold settings. Blue solid curve represents the ROC curve while red dotted line represents random guessing. The greater the Area Under Curve (AUC) or higher ROC score indicates the model has better performance in distinguish different classes.

### Support Vector Machine (SVM)

````python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform

# Initialize the model
svm = SVC(probability=True)  # probability=True is required for ROC-AUC

# Define parameter distribution to search
param_dist_svm = {
    'C': uniform(0.1, 10),
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto'],
    'max_iter': [100, 200, 300]
}

# Use RandomizedSearchCV to perform hyperparameter tuning
random_search_svm = RandomizedSearchCV(estimator=svm, param_distributions=param_dist_svm, scoring='accuracy', n_iter=20, cv=3, verbose=1, random_state=42)
random_search_svm.fit(X_train, y_train)

# Print the best parameters combination
print("Best parameters found:", random_search_svm.best_params_)

# Print the best score (based on accuracy of the model)
print("Best score found:", random_search_svm.best_score_)

print("Best score found:", random_search_svm.best_score_)
````

For SVM, we opt for RandomizedSearchCV instead of GridSearchCV. Here's why:

* RandomizedSearchCV is often used as an efficient alternative to GridSearchCV. While GridSearchCV exhaustively considers all parameter combinations, RandomizedSearchCV samples a given number of candidates from a parameter space with a specified distribution.
* RandomizedSearchCV can often find a good (if not the best) set of hyperparameters in less time than GridSearchCV.
* This approach reduces computational cost, which can be particularly beneficial when dealing with a large number of hyperparameters or when each model training and validation is time-consuming, which is true for SVM (approximate GridSearchCV completion time is 1 hour).

````python
from sklearn.metrics import accuracy_score, f1_score, recall_score

# Function to compute metrics for a model
def compute_metrics(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred, average='macro')
    recall = recall_score(y_true, y_pred, average='macro')
    return accuracy, f1, recall

# Compute metrics
accuracy_logreg, f1_logreg, recall_logreg = compute_metrics(y_test, lr_pred)
accuracy_xgboost, f1_xgboost, recall_xgboost = compute_metrics(y_test, xgb_pred)
accuracy_rf, f1_rf, recall_rf = compute_metrics(y_test, rf_pred)
accuracy_dt, f1_dt, recall_dt = compute_metrics(y_test, dt_pred)
accuracy_ada, f1_ada, recall_ada = compute_metrics(y_test, ada_pred)
accuracy_svm, f1_svm, recall_svm = compute_metrics(y_test, svm_pred)
accuracy_knn, f1_knn, recall_knn = compute_metrics(y_test, knn_pred)
accuracy_autosk, f1_autosk, recall_autosk = compute_metrics(y_test, automl_pred)
````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have computed metrics for these models
model_names = ['logreg', 'XGBoost', 'RandomForest', 'DecisionTree', 'AdaBoost', 'SVM', 'KNN', 'AutoSK']
accuracies = [accuracy_logreg, accuracy_xgboost, accuracy_rf, accuracy_dt, accuracy_ada, accuracy_svm, accuracy_knn, accuracy_autosk]
f1_scores = [f1_logreg, f1_xgboost, f1_rf, f1_dt, f1_ada, f1_svm, f1_knn, f1_autosk]
recalls = [recall_logreg, recall_xgboost, recall_rf, recall_dt, recall_ada, recall_svm, recall_knn, recall_autosk]

# Create a DataFrame
metrics_df = pd.DataFrame({
    'Accuracy': accuracies,
    'F1 Score': f1_scores,
    'Recall': recalls
}, index=model_names)


# Round the DataFrame to 4 decimal places
metrics_df = metrics_df.round(4)
````

````python
# Set the style
sns.set(style='whitegrid')

# Create a heatmap
plt.figure(figsize=(10, 6))
heatmap = sns.heatmap(metrics_df, annot=True, cmap='coolwarm', linewidths=0.5, cbar=True, fmt='.4f')

# Adjustments to improve the layout
plt.title('Model Comparison')
plt.xlabel('Metrics')
plt.ylabel('Models')
plt.show()
````
